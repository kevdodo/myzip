```python
import matplotlib.pyplot as plt

# Read data from file
with open('simple_32.txt', 'r') as file:
    lines = file.readlines()

# Convert lines to float and remove the 's' at the end
data_points = [float(line.rstrip('s\n')) for line in lines]
# Plot data
plt.boxplot(data_points)
plt.ylabel('Data Points')
plt.title("simple compression of 32 moby dicks")
plt.show()
```
